```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib as mb
```

```python
def sigmoid_function(z):
  return 1.0/(1.0-np.exp(z))
```

```python
def calculate_gradient(theta,X,y):
  m = y.size
  return (X.T @(sigmoid_function (X @ theta)- y) / m)
```

```python
def gradient_descent(X,y, alpha = 0.1,num_iter = 100,tol = 1e-7):
   X_b = np.c_[np.ones((X.shape[0],1)), X]
   theta = np.zeros(X_b.shape[1])

   for i in range(num_iter):
    grad = calculate_gradient(theta,X_b,y)
    theta -= alpha * grad

    if np.linalg.norm(grad) < tol:
        break

   return theta
```

```python
def predict_proba(X,theta):
  X_b = np.c_[np.ones((X.shape[0],1)),X]
  return sigmoid_function(X_b@theta)

def predict(X,theta,threshold = 0.5):
  return (predict_proba(X,theta) >= threshold).astype(int)


from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

scaler = StandardScaler()

X,y = load_breast_cancer(return_X_y=True)
X_train,X_test,y_test,y_train = train_test_split(X, y , test_size = 0.2)

X_train_selected = scaler.fit_transform(X_train)
X_test_selected = scaler.fit_transform(X_test)

theta_hat = gradient_descent(X_train_selected,y_train,alpha = 0.1)

y_train_pred = predict(X_train_selected,theta_hat)
y_test_pred = predict(X_test_selected,theta_hat)

train_acc = accuracy_score(y_train,y_train_pred)
test_acc = accuracy_score(y_train,y_test_pred)

print(train_acc)
print(test_acc)
```
